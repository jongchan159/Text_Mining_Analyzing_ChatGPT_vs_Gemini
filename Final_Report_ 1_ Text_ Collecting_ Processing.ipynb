# 2. Stemming (어간 추출) & Lemmatization (표제어 추출)

```python
import pandas as pd
import nltk, re
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords

# --- NLTK 리소스 안전 설치 ---
def safe_nltk_download():
    packs = ["punkt", "punkt_tab", "wordnet", "omw-1.4", "stopwords"]
    for p in packs:
        try:
            nltk.data.find({
                "punkt": "tokenizers/punkt",
                "punkt_tab": "tokenizers/punkt_tab/english.pickle",
                "wordnet": "corpora/wordnet",
                "omw-1.4": "corpora/omw-1.4",
                "stopwords": "corpora/stopwords"
            }[p])
        except LookupError:
            nltk.download(p, quiet=True)

safe_nltk_download()

# --- 데이터 로드 ---
gemini_df_4 = review_df_gemini[review_df_gemini['star']==4]
gemini_df_1 = review_df_gemini[review_df_gemini['star']==1]
gpt_df_4 = review_df_gpt[review_df_gpt['star']==4]
gpt_df_1 = review_df_gpt[review_df_gpt['star']==1]

text_col    = "contents"
```

## 전처리 & 추출 함수

```python
# --- 전처리 ---
def preprocess(text: str) -> str:
    text = str(text).lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# --- 토큰화: NLTK 우선, 실패 시 정규식 우회 ---
def tokenize(text: str):
    try:
        from nltk.tokenize import word_tokenize
        return word_tokenize(text)
    except LookupError:
        # punkt/punkt_tab 설치 이슈 시 우회
        return re.findall(r"[a-z]+", text)

ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
try:
    stop_words = set(stopwords.words("english"))
except LookupError:
    stop_words = set()

def stem_and_lemmatize(text: str):
    toks = [w for w in tokenize(text) if w and w not in stop_words]
    stems  = [ps.stem(w) for w in toks]
    lemmas = [lemmatizer.lemmatize(w) for w in toks]
    return " ".join(stems), " ".join(lemmas)

def process_df(df: pd.DataFrame, text_col: str) -> pd.DataFrame:
    # 슬라이스 경고 방지: 항상 복사본 생성
    out = df.copy()

    # 텍스트 컬럼 없으면 에러
    if text_col not in out.columns:
        raise KeyError(f"텍스트 컬럼 '{text_col}' 이(가) DataFrame에 없습니다. 현재 컬럼: {list(out.columns)}")

    # 전처리
    out.loc[:, "clean_text"] = out[text_col].map(preprocess)

    # (stem, lemma) 튜플 시리즈 -> 2열 DF로 변환 후 붙이기
    pairs = out["clean_text"].apply(stem_and_lemmatize)
    pair_df = pd.DataFrame(pairs.tolist(), columns=["stemmed", "lemmatized"], index=out.index)

    # 병합
    out = pd.concat([out, pair_df], axis=1)
    return out
```

```python
gemini_df_4 = process_df(gemini_df_4, text_col)
gpt_df_4    = process_df(gpt_df_4, text_col)
gemini_df_1 = process_df(gemini_df_1, text_col)
gpt_df_1    = process_df(gpt_df_1, text_col)

# 저장
gemini_df_4.to_csv("/content/drive/MyDrive/AI_media_Text/최종리포트/gemini_processed_4.csv", index=False)
gpt_df_4.to_csv   ("/content/drive/MyDrive/AI_media_Text/최종리포트/gpt_processed_4.csv",    index=False)
gemini_df_1.to_csv("/content/drive/MyDrive/AI_media_Text/최종리포트/gemini_processed_1.csv", index=False)
gpt_df_1.to_csv   ("/content/drive/MyDrive/AI_media_Text/최종리포트/gpt_processed_1.csv",    index=False)
```
