# 2. Stemming (ì–´ê°„ ì¶”ì¶œ) & Lemmatization (í‘œì œì–´ ì¶”ì¶œ)

```python
import pandas as pd
import nltk, re
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords

# --- NLTK ë¦¬ì†ŒìŠ¤ ì•ˆì „ ì„¤ì¹˜ ---
def safe_nltk_download():
    packs = ["punkt", "punkt_tab", "wordnet", "omw-1.4", "stopwords"]
    for p in packs:
        try:
            nltk.data.find({
                "punkt": "tokenizers/punkt",
                "punkt_tab": "tokenizers/punkt_tab/english.pickle",
                "wordnet": "corpora/wordnet",
                "omw-1.4": "corpora/omw-1.4",
                "stopwords": "corpora/stopwords"
            }[p])
        except LookupError:
            nltk.download(p, quiet=True)

safe_nltk_download()

# --- ë°ì´í„° ë¡œë“œ ---
gemini_df_4 = review_df_gemini[review_df_gemini['star']==4]
gemini_df_1 = review_df_gemini[review_df_gemini['star']==1]
gpt_df_4 = review_df_gpt[review_df_gpt['star']==4]
gpt_df_1 = review_df_gpt[review_df_gpt['star']==1]

text_col    = "contents"
```

## ì „ì²˜ë¦¬ & ì¶”ì¶œ í•¨ìˆ˜

```python
# --- ì „ì²˜ë¦¬ ---
def preprocess(text: str) -> str:
    text = str(text).lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# --- í† í°í™”: NLTK ìš°ì„ , ì‹¤íŒ¨ ì‹œ ì •ê·œì‹ ìš°íšŒ ---
def tokenize(text: str):
    try:
        from nltk.tokenize import word_tokenize
        return word_tokenize(text)
    except LookupError:
        # punkt/punkt_tab ì„¤ì¹˜ ì´ìŠˆ ì‹œ ìš°íšŒ
        return re.findall(r"[a-z]+", text)

ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
try:
    stop_words = set(stopwords.words("english"))
except LookupError:
    stop_words = set()

def stem_and_lemmatize(text: str):
    toks = [w for w in tokenize(text) if w and w not in stop_words]
    stems  = [ps.stem(w) for w in toks]
    lemmas = [lemmatizer.lemmatize(w) for w in toks]
    return " ".join(stems), " ".join(lemmas)

def process_df(df: pd.DataFrame, text_col: str) -> pd.DataFrame:
    # ìŠ¬ë¼ì´ìŠ¤ ê²½ê³  ë°©ì§€: í•­ìƒ ë³µì‚¬ë³¸ ìƒì„±
    out = df.copy()

    # í…ìŠ¤íŠ¸ ì»¬ëŸ¼ ì—†ìœ¼ë©´ ì—ëŸ¬
    if text_col not in out.columns:
        raise KeyError(f"í…ìŠ¤íŠ¸ ì»¬ëŸ¼ '{text_col}' ì´(ê°€) DataFrameì— ì—†ìŠµë‹ˆë‹¤. í˜„ì¬ ì»¬ëŸ¼: {list(out.columns)}")

    # ì „ì²˜ë¦¬
    out.loc[:, "clean_text"] = out[text_col].map(preprocess)

    # (stem, lemma) íŠœí”Œ ì‹œë¦¬ì¦ˆ -> 2ì—´ DFë¡œ ë³€í™˜ í›„ ë¶™ì´ê¸°
    pairs = out["clean_text"].apply(stem_and_lemmatize)
    pair_df = pd.DataFrame(pairs.tolist(), columns=["stemmed", "lemmatized"], index=out.index)

    # ë³‘í•©
    out = pd.concat([out, pair_df], axis=1)
    return out
```

```python
gemini_df_4 = process_df(gemini_df_4, text_col)
gpt_df_4    = process_df(gpt_df_4, text_col)
gemini_df_1 = process_df(gemini_df_1, text_col)
gpt_df_1    = process_df(gpt_df_1, text_col)

# ì €ì¥
gemini_df_4.to_csv("/content/drive/MyDrive/AI_media_Text/ìµœì¢…ë¦¬í¬íŠ¸/gemini_processed_4.csv", index=False)
gpt_df_4.to_csv   ("/content/drive/MyDrive/AI_media_Text/ìµœì¢…ë¦¬í¬íŠ¸/gpt_processed_4.csv",    index=False)
gemini_df_1.to_csv("/content/drive/MyDrive/AI_media_Text/ìµœì¢…ë¦¬í¬íŠ¸/gemini_processed_1.csv", index=False)
gpt_df_1.to_csv   ("/content/drive/MyDrive/AI_media_Text/ìµœì¢…ë¦¬í¬íŠ¸/gpt_processed_1.csv",    index=False)
```
